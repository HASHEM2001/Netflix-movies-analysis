```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# Start coding!
```

```python
netflix_df= pd.read_csv('E:\\hashem\\projects_data\\project_one\\netflix_data.csv')
netflix_subset = netflix_df[netflix_df['type']=='Movie']
```

```python
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
short_movies=netflix_movies[netflix_movies['duration']<60]
```

```python
colors=[]
for lab, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        colors.append('red')
    elif row['genre'] == "Documentaries":
        colors.append('black')
    elif row['genre'] == "Stand-Up":
        colors.append('blue')
    else:
        colors.append('green')
```

```python
# Initialize a matplotlib figure object
fig, ax = plt.subplots(figsize=(10, 6))

# Create the scatter plot
ax.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
# Set the labels and title
ax.set_xlabel('Release year')
ax.set_ylabel('Duration (min)')
ax.set_title('Movie Duration by Year of Release')
plt.show()
```

```python
# Create a figure and a 1x2 grid of subplots (2 subplots side by side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))  # Width and height in inches

# Plot on the first subplot
ax1.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)
ax1.set_xlabel('Release Year')
ax1.set_ylabel('Duration (min)')
ax1.set_title('Movie Duration by Year of Release')

# Plot on the second subplot
ax2.hist(netflix_movies['duration'], bins=5, color='blue')
ax2.set_xlabel('Duration (min)')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Movie Duration')

# Adjust layout
plt.tight_layout()

# Display the pl
plt.show()
```
